```python
import pandas as pd
import numpy as np

#reading the data file!!
df=pd.read_csv('data.csv')

df
```

```python
df.info()
```

```python
# The id column is not useful!
df.drop('id', axis=1, inplace=True)

df.drop(['Unnamed: 32'], axis=1, inplace=True)

# Converting the Diagnosis parameter into integer
df['diagnosis']=df['diagnosis'].str.replace('M','0')
df['diagnosis']=df['diagnosis'].str.replace('B','1')
df['diagnosis']=df['diagnosis'].astype('int32')

#values = {'M':0, 'B':1}
# Map the dictionary to the column
#df['diagnosis'] = df['diagnosis'].map(values)
#df['diagnosis'].unique()
df.isnull().sum()
```

```python
import plotly.express as px

#px.imshow(df.corr().values,x=df.corr().index,y=df.corr().columns())

corr_matrix = df.corr()

# Create a correlation plot using plotly
fig = px.imshow(corr_matrix.values, x=corr_matrix.index, y=corr_matrix.columns,color_continuous_scale='Viridis')

fig.update_layout(
    title="Correlation Plot",
    xaxis_title="Features",
    yaxis_title="Features"
)

```

```python
# Calculate correlations with the target variable
correl=df.corr()['diagnosis']

# Filter correlations based on the threshold (0.3)
significant_corr= correl[abs(correl) > 0.2]

# Print the significant correlations
print("Significant Correlations:")
print(significant_corr)

```

```python
input=df.drop('diagnosis', axis=1)
output=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split as tt

X_train,X_test,y_train,y_test=tt( input, output, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

log_r=LogisticRegression()
```

```python
#logistics
log_rm=log_r.fit(X_train, y_train)
log_r_preds=log_rm.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Prints model results
def ModelStats(ground_truths, predictions):
  acc = accuracy_score(ground_truths, predictions)

  # Model performance with true positive (tp), false positive (fp), false negative (fn) and true negative (tn)
  cm = confusion_matrix(ground_truths, predictions)

  print(f'Accuracy: {acc}\n')
  print(f'Confusion Matrix:\n{cm}')
```

```python
ModelStats(y_test, log_r_preds)
```
